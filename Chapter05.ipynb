```python
import geopandas as gpd
import fiona

fiona.drvsupport.supported_drivers['LIBKML'] = 'rw'
data = gpd.read_file('chapter_05_data\ParisSeineData.kml')
print(data)
```

```python
seine = data.iloc[0:1,:]
seine.plot()
```

```python
paris = data.iloc[1:2,:]
paris.plot()
```

```python
paris_seine = seine.clip(paris)
paris_seine
```

```python
paris_seine.plot()
```

```python
fiona.drvsupport.supported_drivers['KML'] = 'rw'
data = gpd.read_file('chapter_05_data\ParisSeineData_example2.kml')
data.head()
```

```python
data.plot(cmap='tab10')
```

```python
# Extract periph data
periph = data.iloc[0:1,:]

# Extract seine data
seine = data.iloc[1:2,:]

intersection = seine.overlay(periph, how='intersection')
intersection
```

```python
intersection.plot()
```
